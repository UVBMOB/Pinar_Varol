```python
import numpy as np
import scipy as st
from scipy.stats import laplace
import seaborn as sns
import matplotlib.pyplot as plt

class LaplaceDist:
    def __init__(self, n, mean, stddev):
        self.n = n
        self.mean = mean
        self.stddev = stddev
    def createDist(self):
        self.Dist=laplace.rvs(loc=self.mean,scale=self.stddev,size=self.n)
    def calculateCdf(self):
        self.calculateCdf=laplace.cdf(self.Dist)
        sns.lineplot(x=self.Dist, y=self.calculateCdf)
        plt.show()
        return
    def calculatePdf(self):
        self.calculatePdf=laplace.pdf(self.Dist)
        sns.lineplot(x=self.Dist, y=self.calculatePdf)
        plt.show()
        return
```

```python
ornek1=LaplaceDist(10000,0,1)
ornek1.createDist()
ax=sns.distplot(ornek1.Dist)
```
